```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'/content/drive/MyDrive/Stock Market Sentiment Analysis/Data.csv',encoding='ISO-8859-1')
df.head()
```

```python
train=df[df['Date']<'20150101']
```

```python
test= df[df['Date']>'20141231']
```

```python
train.head()
```

```python
data=train.iloc[:,2:27]
```

```python
data.head()
```

```python
data.columns
```

```python
data.replace('[^a-zA-Z]',' ',inplace=True)
```

```python
data.head()
```

```python
new_index=[str(i) for i in range(25)]
new_index
```

```python
data.columns=new_index
data.head()
```

```python
data.index
```

```python
#lower case conversion

for index in new_index:
  data[index]=data[index].str.lower()

data.head()
```

```python
# performing data wrangling and data merging
data.iloc[1,0:25]
```

```python
headlines=[]
for i in data.iloc[1,0:25]:
  headlines.append(i)

' '.join(headlines)
```

```python
' '.join([str(i) for i in data.iloc[1,0:25]])
```

```python
headlines=[]
for row in range(0,len(data)):
  headlines.append(' '.join([str(i) for i in data.iloc[row,0:25]]))
```

```python
headlines[0:3]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
cv=CountVectorizer(ngram_range=(2,2))
```

```python
traindata_x=cv.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
```

```python
randomclassifier.fit(traindata_x,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test)):
  test_transform.append(' '.join([str(i) for i in test.iloc[row,2:27]]))
```

```python
test_data=cv.transform(test_transform)
```

```python
prediction=randomclassifier.predict(test_data)
prediction
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(test['Label'],prediction)
cm
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(cm,cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
labels=['positive','negative']
tick_marks=np.arange(len(labels))
plt.xticks(tick_marks,labels)
plt.yticks(tick_marks,labels)
plt.tight_layout()
plt.xlabel('True Labels')
plt.ylabel('Predicted Labels')
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(test['Label'],prediction)
```

```python
report=classification_report(test['Label'],prediction)
```

```python
print(report)
```

```python
def plot_confusion_matrix(cm,title='Confusion Matrix'):
  plt.imshow(cm,cmap=plt.cm.Blues)
  plt.title('Confusion Matrix')
  plt.colorbar()
  labels=['positive','negative']
  tick_marks=np.arange(len(labels))
  plt.xticks(tick_marks,labels)
  plt.yticks(tick_marks,labels)
  plt.tight_layout()
  plt.xlabel('True Labels')
  plt.ylabel('Predicted Labels')

```

```python
plot_confusion_matrix(cm1)
```
